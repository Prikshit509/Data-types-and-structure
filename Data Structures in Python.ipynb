For Loop

Expected Output: 10 20 30 40 50 60 70

Approach 1: without index numbers, how can we directly access the elements

```python
lst = [10, 20, 30, 40, 50, 60, 70]
for num in lst:
  print(num, end = " ")
```

Size of the list => Number of elements in the list

```python
len(lst)
```

Approach 2: with the index numbers, we can directly access the elements

```python
for i in range(len(lst)):
  print(lst[i], end = " ")
```

```python
for i in range(0,len(lst), 2):
  print(lst[i], end = " ")
```

```python
for i in range(1,5):
  print(lst[i], end = " ")
```

while loop

```python
num = 1

while num < 5:
  print(num)
  num = num + 1
```

```python
lst1 = [10, 12, 14, 15, 17, 21, 22, 25, 27, 29]
## write a code to print even elements and odd elements in the output
for num in lst1:
  ## number is a even number
  if num % 2 == 0:
    print(f"{num} is an even number")
  else:
    print(f"{num} is a odd number")
```
